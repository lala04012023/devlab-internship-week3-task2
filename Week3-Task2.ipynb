```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Lala\Downloads\archive\hotel_bookings.csv")
```

```python
df
```

```python
df.isnull().sum()
```

## Boş dəyərlərin müvafiq verilənlə doldurulması

```python
df['country']=df['country'].fillna('Unknown')
```

```python
df['agent']=df['agent'].fillna(0)
```

```python
df['company']=df['company'].fillna(0)
```

```python
df.isnull().sum()
```

## arrival date sütünunun yaradılması

```python
df['arrival_date']=pd.to_datetime(df['arrival_date_year'].astype(str)+' '+ df['arrival_date_month']+' '+ df['arrival_date_day_of_month'].astype(str))
```

```python
df
```

## Pivot Table yaradılması

```python
pivot_table=pd.pivot_table(df,index='arrival_date_day_of_month', columns='hotel', values='is_canceled',aggfunc='count')
```

```python
pivot_table
```

## Cancel_rate

```python
cancel_rate= df.groupby(["arrival_date_month", "hotel"])["is_canceled"].mean().reset_index()
```

```python
cancel_rate
```

## Compare average lead_time

```python
df.groupby("is_canceled")["lead_time"].mean()

```

## TOP10 country

```python
top10 = (df.groupby("country").agg(
        total_bookings=("country", "count"),
        cancellation_rate=("is_canceled", "mean") ) .sort_values("total_bookings", ascending=False).head(10))
```

```python
top10
```

## Analyze demand by market_segment 

```python
market_segment = ( df.groupby("market_segment") .agg(
        total_bookings=("market_segment", "count"),
        cancellation_rate=("is_canceled", "mean")).sort_values("total_bookings", ascending=False))
```

```python
market_segment
```

### 1)Booking demand is highly seasonal, with some months showing significantly higher reservation volumes than others. Revenue management strategies such as dynamic pricing and staffing adjustments should focus on these peak periods to maximize occupancy and profitability.
### 2)Guests who book much earlier (higher lead_time) tend to cancel more frequently than those who make reservations closer to their arrival date. Hotels should consider flexible pricing strategies or partial prepayment requirements for long lead-time bookings to reduce cancellation losses.
### 3)The majority of reservations come from a limited number of countries. Monitoring cancellation behavior in these key markets allows hotels to design country-specific promotional campaigns and reservation policies that improve booking retention and reduce revenue uncertainty.
### 4)Online Travel Agencies (Online TA) generate the largest share of bookings, making them the most important acquisition channel. However, if this segment also exhibits a relatively high cancellation rate, hotels should encourage more direct bookings through loyalty programs, exclusive discounts, or value-added benefits to improve customer retention and reduce commission expenses.

# BONUS 1:

```python
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December']

df['arrival_date_month'] = pd.Categorical(
    df['arrival_date_month'],
    categories=month_order,
    ordered=True)


adr = df.groupby(['arrival_date_month', 'hotel'])['adr'].mean().unstack()
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(
    adr.index,
    adr['City Hotel'],
    marker='o',
    linewidth=2,
    label='City Hotel',
    color='blue')
ax1.set_xlabel('Month')
ax1.set_ylabel('City Hotel ADR', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.tick_params(axis='x', rotation=45)

ax2 = ax1.twinx()

ax2.plot(
    adr.index,
    adr['Resort Hotel'],
    marker='s',
    linewidth=2,
    label='Resort Hotel',
    color='red')
ax2.set_ylabel('Resort Hotel ADR', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Average Daily Rate (ADR) by Month and Hotel Type')
plt.show()
```

### Resort Hotel-də orta gündəlik otaq qiyməti (ADR) əsasən yay və tətil mövsümlərində daha yüksək olur.
### City Hotel-də ADR il ərzində daha sabit qalır.
### Mövsümi tələb Resort Hotel-in qiymətlərinə City Hotel ilə müqayisədə daha çox təsir göstərir.
